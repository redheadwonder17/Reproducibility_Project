R Code pulled from https://www.statmethods.net/advgraphs/ggplot2.html

```r
df = read.delim("https://doi.org/10.1371/journal.pone.0196058.s002")
```

```r
library(tidyverse)
ggplot(data = df, aes(x = Het_Hom, y = Ti_Tv, color = StudyType)) +
  geom_point() +
  ylab("PPV (Ti/Tv)") + 
  xlab("Het/Hom") +
  theme(panel.grid.major = element_line(color="grey", size = (1.0)),
                   panel.grid.minor = element_line(size = (0.2), color="grey"))
```
